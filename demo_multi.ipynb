```python
# Run predictions on the test data and store the output labels.
# Used to calculate performance metrics in the next cell
from utils import getVideoFilesFromFolder
from combined_model import predict

dir_paths = ['./dataset/audiovisual_test/anger/',
             './dataset/audiovisual_test/disgust/',
             './dataset/audiovisual_test/fear/',
             './dataset/audiovisual_test/happiness/',
             './dataset/audiovisual_test/sadness/',
             ]
labels = ['anger', 'disgust', 'fear', 'happiness', 'sadness', 'surprise', 'neutral']
debug=False

y_pred_meta = []
y_pred_max = []
y_pred_aud = []
y_pred_vis = []
y_true = []
for dir_path, label in zip(dir_paths, labels):
    files = getVideoFilesFromFolder(dir_path)
    for i, f in enumerate(files):
        print(f"Predicting {i} of {len(files)} files")
        pred = predict(multimodal, f)
        y_pred_aud.append(pred[2])
        y_pred_vis.append(pred[3])
        y_pred_max.append(pred[1])
        y_pred_meta.append(pred[0][0])
        y_true.append(label)

print("Test run completed")
```

```python
from sklearn.metrics import confusion_matrix, f1_score, precision_recall_fscore_support, accuracy_score

print("Consfusion meta clf")
print(confusion_matrix(y_true, y_pred_meta, labels=labels))
print(f1_score(y_true, y_pred_meta, average='weighted'))
print(accuracy_score(y_true, y_pred_meta))
precision, recall, fscore, support  = precision_recall_fscore_support(y_true, y_pred_meta)
print('precision: {}'.format(precision))
print('recall: {}'.format(recall))
print('fscore: {}'.format(fscore))
print('support: {}'.format(support))

print("Consfusion aud")
print(confusion_matrix(y_true, y_pred_aud))
print(f1_score(y_true, y_pred_aud, average='weighted'))
print(accuracy_score(y_true, y_pred_aud))
precision, recall, fscore, support  = precision_recall_fscore_support(y_true, y_pred_aud)
print('precision: {}'.format(precision))
print('recall: {}'.format(recall))
print('fscore: {}'.format(fscore))
print('support: {}'.format(support))

print("Consfusion vis")
print(confusion_matrix(y_true, y_pred_vis, labels=labels))
print(f1_score(y_true, y_pred_vis, average='weighted'))
print(accuracy_score(y_true, y_pred_vis))
precision, recall, fscore, support  = precision_recall_fscore_support(y_true, y_pred_vis)
print('precision: {}'.format(precision))
print('recall: {}'.format(recall))
print('fscore: {}'.format(fscore))
print('support: {}'.format(support))
```

```python
from pyAudioAnalysis import MidTermFeatures as mT

dir_paths = ['./dataset/audiovisual/anger/',
             './dataset/audiovisual/disgust/',
             './dataset/audiovisual/fear/',
             './dataset/audiovisual/happiness/',
             './dataset/audiovisual/sadness/',
             ]
out = mT.multiple_directory_feature_extraction(dir_paths, 1, 1, 0.05, 0.05)
```

```python
# Perform PCA and display most important dimensions

import pandas as pd
import numpy as np

feats = out[0]
labels = ['anger', 'disgust', 'fear', 'happiness', 'sadness']

cols = []
for feat, lab in zip(feats, labels):
    cols.extend([lab]*len(feat))

feats = np.concatenate(feats)

from sklearn.decomposition import PCA

pca = PCA(n_components=6)
X = pca.fit_transform(feats)


import plotly.express as px
px.scatter_matrix(X, dimensions=[0,1,2], color=cols).show()
```

```python
# Histogram of the most discriminative feature

import plotly.figure_factory as ff
df = pd.DataFrame(feats)
df['labels'] = cols

active_labels = ['anger', 'disgust', 'sadness']

x = []
i=78
for lab in active_labels:
    x1 = df.loc[df.labels==lab, i]
    x.append(x1)

fig = ff.create_distplot(x, active_labels, bin_size=.02)
fig.update_layout({'title': 'mfcc_3_std'})
fig.show()
```
